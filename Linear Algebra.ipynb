```python
import numpy as np
import matplotlib.pyplot as plt
```

# 1. 벡터 공간

- 벡터
: 크기와 방향을 가진 물리량



- 벡터의 합
: 벡터 합의 평행사변형 법칙

```python
# 두 개의 벡터 생성
vector1 = np.array([1, 3])
vector2 = np.array([3, 1])

# 벡터의 합 계산
vector_sum = vector1 + vector2

# 원점에서 벡터를 시작하여 그래프로 표시
plt.quiver(0, 0, vector1[0], vector1[1], angles='xy', scale_units='xy', scale=1, color='r', label='Vector 1')
plt.quiver(vector1[0], vector1[1], vector2[0], vector2[1], angles='xy', scale_units='xy', scale=1, color='b', label='Vector 2')
plt.quiver(0, 0, vector_sum[0], vector_sum[1], angles='xy', scale_units='xy', scale=1, color='g', label='Vector Sum')

# 축 범위 설정
plt.xlim(0,  4)
plt.ylim(0, 4)

# 축 레이블 설정
plt.xlabel('X')
plt.ylabel('Y')

# 그리드 추가
plt.grid()

# 벡터와 벡터 합 표시
plt.legend()

# 그래프 표시
plt.show()
```

 - 스칼라곱
 : 벡터에 실수를 곱해 크기를 확대하거나 축소할 수 있다.

```python
x = np.array([1, 2])
t = 2
result = x * t
plt.quiver(0, 0, x[0], x[1], angles = 'xy', scale_units='xy', scale=1, color='r', label='Vector 1')
plt.quiver(x[0], x[1], result[0], result[1], angles = 'xy', scale_units='xy', scale=1, color='b', label='Vector 1-Scalar Multiplication')
plt.xlim(0, 6)
plt.ylim(0, 12)
plt.xlabel('X')
plt.ylabel('Y')


# 벡터와 벡터 합 표시
plt.legend(loc ='lower right')

# 그래프 표시
plt.show()
```

- 벡터의 합과 스칼라 곱의 8가지 성질
    - 성질 1 : 모든 벡터 x, y에 대하여 x + y = y + x이다.
    - 성질 2 : 모든 벡터 x, y, z에 대하여 (x+y)+z = x + (y+z)이다.
    - 성질 3 : 모든 벡터 x에 대하여 x + 0 = x를 만족하는 벡터 O이 존재한다.
    - 성질 4 : 각 벡터 x마다 x + y = O을 만족하는 벡터 y가 존재한다.
    - 성질 5 : 모든 벡터 x에 대하여 1x = x이다.
    - 성질 6 : 모든 실수 a, b와 모든 벡터 x에 대하여 (ab)x = a(bx)이다.
    - 성질 7 : 모든 실수 a와 모든 벡터 x, y에 대하여 a(x + y) = ax + ay이다.
    - 성질 8 : 모든 실수 a, b와 모든 벡터 x에 대하여 (a + b)x = ax + bx이다.
